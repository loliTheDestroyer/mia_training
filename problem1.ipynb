```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('Weather Dataset.csv')
```

```python
df.describe()
```

```python
df.drop_duplicates(subset='Formatted Date', inplace=True)
df.describe()
```

```python
df.dropna(inplace=True)
df.describe()
```

```python
df.drop(columns='Loud Cover', inplace=True, )
df.describe()
```

```python
df=df[(df['Temperature (C)']<=50) & (df['Temperature (C)']>=-30)
      & (df['Visibility (km)']>=0) &
      (df['Pressure (millibars)']<=1050) & (df['Pressure (millibars)']>=950)
      ]
df.describe()
```

```python
plt.figure(figsize=(15 ,10))
plt.plot(df['Formatted Date'], df['Temperature (C)'])
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Time')
plt.show()
```

```python

plt.hist(np.random.normal(df['Temperature (C)'].mean(), df['Temperature (C)'].std(), df.shape[0]))
plt.title('temperature distribution')
plt.show()
```

```python
plt.scatter(df['Temperature (C)'], df['Humidity'])
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title('Temperature vs Humidity')
plt.show()
```

```python
plt.figure(figsize=(10 ,10))
corr = df.select_dtypes(include=['float64', 'int64']).corr()
sb.heatmap(corr, annot = True, cmap='coolwarm')
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.show()
```

The map explains the correlation between each 2 variables, 1 is perfect correlation, when one variable changes, the other changes in the same direction. -1 represents negative perfect correlation, when one changes, the other changes in the opposit direction. values between 0.6 and 1. while 0 is no correlation.
For example Temperature and pressure have correlation coefficient of -0.64, pressure has moderate tendency to increse when temperature decreases, and vice versa.
the diagonal in red is the correlation between each variable and itself which is always 1.
another example is pressure and windbearing, with correlation coefficient of   0.067, the one has almost no tendency to change when the other changes.
